```python
import pandas as pd
```

```python
df = pd.read_csv('data/grades.csv')
df.head(2)
```

```python
df1 = df.iloc[:,[0,1,3,5]]
df1.head(2)
```

```python
# Series for each column
df1.apply(lambda z: print(type(z)))
```

```python
df1.apply(lambda z: print(len(z))) # each z is a column
```

```python
df1.apply(lambda row: print(row[0]), axis = 1)
df1.apply(lambda row: print(row[1]), axis = 1)
```

```python
df1['Final'] = df1.apply(lambda row: (row[1] + row[2] + (2*row[3])) / 400, axis = 1) 
# each student's average, with final exam weighting double
# we are iterating over rows, where indeces are the columns
df1.head(2)
```

```python
# Allow top 1% of students to go to a party
num_students = int(len(df) * 0.01)
num_students
```

```python
# IDs of 23 top 1% students
indicies = df1.Final.argsort() # argsort sorts in ascending order
indicies.head()
```

```python
df1.Final[1424] # score of the lowest scoring person
```

```python
filtered_students = df1.iloc[indicies][-num_students:][::-1] # this will show me sorted indices for the top 23 students!
# [::-1] reverses the list

filtered_students.student_id.values # this creates an array of the students that will be selected to go to the party
```
